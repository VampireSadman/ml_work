```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("canada_per_capita_income.csv")
df
```

```python
df1=df.rename(columns={
    'per capita income (US$)':'Income'
})
```

```python
df1
```

```python
%matplotlib inline
plt.xlabel("YEAR")
plt.ylabel("INCOME(US$)")
plt.title("Per_capital_income")
plt.scatter(df1.year,df1.Income,color='red',marker='*')
```

```python
new_df=df1.drop('Income',axis='columns')
new_df
```

```python
income=df1.Income
income
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(new_df, income)
```

```python
reg.predict([[2027]])
```

```python
%matplotlib inline
plt.xlabel("YEAR")
plt.ylabel("INCOME(US$)")
plt.title("Per_capital_income")
plt.scatter(df1.year,df1.Income,color='red',marker='*')
plt.plot(df1.year,reg.predict(df1[['year']]),color='blue')
```
